```python
import json
import os

import numpy as np
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
plt.rcParams.update({
    'font.size': 20,              # 默认字体大小
    'axes.titlesize': 24,         # 标题字体大小
    'axes.labelsize': 20,         # 轴标签字体大小
    'xtick.labelsize': 20,        # x轴刻度标签字体大小
    'ytick.labelsize': 20,        # y轴刻度标签字体大小
    'legend.fontsize': 12         # 图例字体大小
})
```

```python
def load_scores(path: str, field: str):
    """
    从 JSON 文件中加载分数字典，然后提取 field 对应的列表。
    """
    if not os.path.exists(path):
        raise FileNotFoundError(f"文件不存在：{path}")
    with open(path, "r") as f:
        data = json.load(f)
    if field not in data:
        raise KeyError(f"字段 {field} 不存在于文件 {path}")
    return np.array(data[field])
```

```python
def plot_platform_roc(platform: str, watermark_dict: dict, key_dict: dict,savefig_path:str):
    """
    platform: 'OPT-6.7B' 或 'GPTJ-6B'
    watermark_dict: watermarks 中对应数据
    key_dict: keys 中对应字段名
    """
    plt.figure(figsize=(8, 6))
    for method_name, files in watermark_dict.items():
        # 1. 加载阳性和阴性分数
        pos_path = files[platform]
        neg_path = files[f"{platform.split('-')[0]}-UNMARK"]
        pos_field = key_dict[method_name][platform]
        neg_field = key_dict[method_name][f"{platform.split('-')[0]}-UNMARK"]

        y_scores_pos = load_scores(pos_path, pos_field)
        y_scores_neg = load_scores(neg_path, neg_field)

        # 2. 对 EXP-Edit 方法做反向处理：正样本分数低 → 取反之后高
        if method_name == "EXP-Edit":
            y_scores_pos = -y_scores_pos
            y_scores_neg = -y_scores_neg

        # 3. 构造标签与得分数组
        y_true = np.concatenate([np.ones_like(y_scores_pos), np.zeros_like(y_scores_neg)])
        y_scores = np.concatenate([y_scores_pos, y_scores_neg])

        # 4. 计算 ROC 曲线和 AUC
        fpr, tpr, _ = roc_curve(y_true, y_scores)
        roc_auc = auc(fpr, tpr)
        if method_name=="kgw":
            method_name="KGW"
        elif method_name=='Ours_KGW1':
            method_name="LTW-1"
        elif method_name=="Ours_KGW0":
            method_name="LTW-0"
        # 5. 绘制
        plt.plot(fpr, tpr, label=f"{method_name} (AUC = {roc_auc:.3f})")

    # 基准对角线
    plt.plot([0, 1], [0, 1], linestyle="--", label="Random Guess")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(f"ROC Curves on {platform}")
    plt.legend(loc="lower right")
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(savefig_path)
    plt.show()
```

```python
# OPT-6.7B 平台
plot_platform_roc("OPT-6.7B", watermarks, keys,'/data3/wcr/my_project/selective_watermark_based_on_semantic/eval/dipper_auroc_opt.pdf')
# GPTJ-6B 平台
plot_platform_roc("GPTJ-6B", watermarks, keys,'/data3/wcr/my_project/selective_watermark_based_on_semantic/eval/dipper_auroc_gptj.pdf')
```
