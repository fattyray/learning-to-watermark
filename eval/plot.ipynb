```python
import json
data={}
def load_json(file_path):
    """从JSON文件加载数据"""
    with open(file_path, 'r') as f:
        return json.load(f)
from sklearn.metrics import roc_auc_score
import numpy as np

def compute_auroc(positive_scores, negative_scores, reverse=False):
    scores = np.concatenate([positive_scores, negative_scores])
    labels = np.concatenate([np.ones(len(positive_scores)), np.zeros(len(negative_scores))])
    if reverse:
        scores = -scores
    return roc_auc_score(labels, scores)


from sklearn.metrics import f1_score

def compute_best_f1_score(positive_scores, negative_scores, reverse=False):
    scores = np.concatenate([positive_scores, negative_scores])
    labels = np.concatenate([np.ones(len(positive_scores)), np.zeros(len(negative_scores))])

    if reverse:
        scores = -scores

    best_f1 = 0.0
    best_threshold = None

    for threshold in np.unique(scores):
        predictions = (scores >= threshold).astype(int)
        f1 = f1_score(labels, predictions)
        if f1 > best_f1:
            best_f1 = f1
            best_threshold = threshold

    return best_f1
```

```python
#eval tpr,auroc,bestf1 score after dipper attack
def compute_tpr_at_fpr( my_wm_z,un_wm_z, fpr, reverse=False):

    sorted_un_wm_z = np.sort(un_wm_z) if reverse else np.sort(un_wm_z)[::-1]  
    threshold = sorted_un_wm_z[int(fpr * len(sorted_un_wm_z))]
    if reverse:
        tp = np.sum(np.array(my_wm_z) < threshold)  
    else:
        tp = np.sum(np.array(my_wm_z) > threshold)
    tpr = tp / len(my_wm_z)  # TPR = TP / (TP + FN)

    
    return tpr
```

```python
def generate_latex_table(data_dict, dipper_data_dict, caption, label):

    # 创建数据框
    df_no_attack = pd.DataFrame.from_dict(data_dict, orient='index')
    df_dipper = pd.DataFrame.from_dict(dipper_data_dict, orient='index')
    
    # 根据要求移除不需要的列
    if 'tpr@1' in df_no_attack.columns:
        df_no_attack = df_no_attack.drop(columns=['tpr@1'])
    
    if 'tpr@5' in df_dipper.columns:
        df_dipper = df_dipper.drop(columns=['tpr@5'])
    
    # 提取文本质量相关列
    text_quality_cols = ['avg_ppl', 'avg_sim']
    df_text_quality = df_no_attack[text_quality_cols].copy()
    
    # 从原始数据中移除文本质量列，只保留评价指标
    df_no_attack = df_no_attack.drop(columns=text_quality_cols)
    
    # 定义列名映射字典
    column_name_map = {
        'best_f1': 'Best F1',
        'avg_ppl': 'Perplexity',
        'avg_sim': 'Semantic Similarity',
        'auroc': 'AUROC',
        'tpr@2': 'TPR@2',
        'tpr@10': 'TPR@10'
    }
    
    # 生成LaTeX代码
    latex_code = "\\begin{table}[htbp]\n"
    latex_code += "\\centering\n"
    latex_code += "\\caption{" + caption + "}\n"
    latex_code += "\\label{" + label + "}\n"
    
    # 计算列数
    no_attack_cols = len(df_no_attack.columns)
    dipper_cols = len(df_dipper.columns)
    text_quality_cols = len(df_text_quality.columns)
    total_cols = no_attack_cols + dipper_cols + text_quality_cols + 1  # +1 for Method column
    
    # 创建表格环境
    latex_code += "\\begin{tabular}{l" + "c" * (total_cols - 1) + "}\n"
    latex_code += "\\toprule\n"
    
    # 创建多级表头
    latex_code += "\\multirow{2}{*}{Method} & \\multicolumn{" + str(no_attack_cols) + "}{c}{No Attack} & "
    latex_code += "\\multicolumn{" + str(dipper_cols) + "}{c}{Dipper Attack} & "
    latex_code += "\\multicolumn{" + str(text_quality_cols) + "}{c}{Text Quality} \\\\\n"
    
    # 添加子表头
    latex_code += " & "
    # 使用映射后的列名
    no_attack_headers = [column_name_map.get(col, col) for col in df_no_attack.columns]
    dipper_headers = [column_name_map.get(col, col) for col in df_dipper.columns]
    text_quality_headers = [column_name_map.get(col, col) for col in df_text_quality.columns]
    
    latex_code += " & ".join(no_attack_headers)
    latex_code += " & "
    latex_code += " & ".join(dipper_headers)
    latex_code += " & "
    latex_code += " & ".join(text_quality_headers)
    latex_code += " \\\\\n"
    
    latex_code += "\\midrule\n"
    
    # 添加数据行
    for method in df_no_attack.index:
        if method in df_dipper.index and method in df_text_quality.index:
            # 将方法名中的下划线替换为连字符
            method_display = method.replace('_', '-')
            row = [method_display]
            
            # 添加No Attack数据
            for col in df_no_attack.columns:
                value = df_no_attack.loc[method, col]
                if pd.isna(value):
                    row.append("--")
                elif col in ['auroc', 'best_f1']:
                    row.append(f"{value:.5f}")
                elif col == 'tpr@2':
                    row.append(f"{value:.3f}")
                else:
                    row.append(f"{value:.5f}")
            
            # 添加Dipper Attack数据
            for col in df_dipper.columns:
                value = df_dipper.loc[method, col]
                if pd.isna(value):
                    row.append("--")
                elif col in ['auroc', 'best_f1']:
                    row.append(f"{value:.5f}")
                elif col == 'tpr@10':
                    row.append(f"{value:.3f}")
                else:
                    row.append(f"{value:.5f}")
            
            # 添加Text Quality数据
            for col in df_text_quality.columns:
                value = df_text_quality.loc[method, col]
                if pd.isna(value):
                    row.append("--")
                elif col == 'avg_ppl':
                    row.append(f"{value:.4f}")
                else:
                    row.append(f"{value:.4f}")
            
            latex_code += " & ".join(row) + " \\\\\n"
    
    latex_code += "\\bottomrule\n"
    latex_code += "\\end{tabular}\n"
    latex_code += "\\end{table}\n"
    
    return latex_code
```
